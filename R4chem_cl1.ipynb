```r
# Crear una lista de 10 números del 1 al 10 usando seq()
mi_lista2 <- seq(1, 10)

# Mostrar la lista creada
print(mi_lista2)
```

```r
# Generar 10 números enteros aleatorios entre 1 y 100
numeros_aleatorios <- sample(1:10, 10, replace = TRUE)

# Mostrar los números aleatorios generados
print(numeros_aleatorios)
```

```r
# Generar 10 números aleatorios de punto flotante entre 0 y 1
numeros_aleatorios <- runif(10)

# Mostrar los números aleatorios generados
print(numeros_aleatorios)
```

```r
# Suponiendo que ya tienes definidas las listas mi_lista y numeros_aleatorios

# Graficar mi_lista versus numeros_aleatorios
plot(mi_lista2, numeros_aleatorios,
     main = "Gráfico de mi_lista versus numeros_aleatorios",
     xlab = "mi_lista", ylab = "numeros_aleatorios",
     col = "red", pch = 2)
```

```r
library(ggplot2)

# Suponiendo que tienes definidas las listas mi_lista y numeros_aleatorios

# Crear un data frame con tus datos
data <- data.frame(
  mi_lista = mi_lista,
  numeros_aleatorios = numeros_aleatorios
)

# Crear el gráfico usando ggplot
ggplot(data, aes(x = mi_lista, y = numeros_aleatorios)) +
  geom_point(color = "blue", size = 5) +
  labs(title = "Gráfico de mi_lista versus numeros_aleatorios",
x = "mi_lista", y = "numeros_aleatorios")
```

## ¿Cómo aplico una función?

```r
# Función para calcular la suma de los primeros n términos de la serie
suma_n_terminos <- function(n) {
  suma <- 0
  for (i in 1:n) {
    suma <- suma + i
  }
  return(suma)
}
```

```r
# Define el número de términos que deseas sumar
n <- 10  # Cambia este valor por la cantidad de términos que desees sumar

# Calcula la suma de los primeros n términos
resultado <- suma_n_terminos(n)
print(resultado)
```
